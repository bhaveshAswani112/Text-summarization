```python
from dotenv import load_dotenv
import os
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage , AIMessage, SystemMessage
load_dotenv()
```

```python
groq_api_key = os.getenv("groq_api_key")
llm = ChatGroq(api_key=groq_api_key)
```

```python
speech = ''' My dear countrymen, my family members!

Today is that auspicious moment when we pay our respects to the innumerable
revered brave freedom fighters who sacrificed their lives for the country and
committed themselves for the freedom of the country; who struggled
throughout their lives, embraced the gallows bravely with slogans of Bharat
Mata ki Jai. It is a festival to remember their grit, resolve and patriotic virtues.
It is due to these bravehearts that we have the good fortune of breathing
freely on this festival of independence. The country is deeply indebted to them.
We express our reverence towards every such great personality.

Dear Countrymen,

Today I extend my deep respect to all those who are protecting the country
with full dedication with commitment for nation building; and are also trying to
take the country to new heights. Be it our farmers, the high morale of the
youth, the contributions of our mothers and sisters; or the dalits, oppressed,
exploited, deprived; today their patriotic zeal and faith in democracy have
become an inspiration for the world. I salute all such people with deep respect.'''
```

```python
chat_message = [
    SystemMessage(content="You are an expert in  summarizing text."),
    HumanMessage(content=f"Summarize the given text : {speech}")
]
```

```python
llm.get_num_tokens(speech)
```

```python
llm(chat_message)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
generic_template = 'You are expert in summarizing the given text\n : {speech}.\n\n. Convert the summary of speech into the {language} language'

prompt = PromptTemplate(
    input_variables = ['speech','language'],
    template = generic_template
)
```

```python
prompt
```

```python
from langchain.chains import LLMChain
```

```python
chain = LLMChain(
    llm=llm,prompt=prompt
)
```

```python
chain.invoke({"speech" : speech , "language" : "hindi"})
```

```python
from langchain.document_loaders import PyPDFLoader
from langchain.chains.summarize import load_summarize_chain
```

```python
generic_template = 'You are expert in summarizing the given text\n : {text}.\n\n. Convert the summary of speech into the {language} language'

prompt = PromptTemplate(
    input_variables = ['text','language'],
    template = generic_template
)
```

```python
loader = PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python
chain = load_summarize_chain(llm=llm,chain_type='stuff',verbose=True,prompt=prompt)
```

```python
chain.invoke({"text" : speech, "language" : "english","input_documents" : docs})
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
docs = RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
```

```python
len(docs)
```

```python
generic_template = 'You are expert in summarizing the given text\n : {text}.'

prompt = PromptTemplate(
    input_variables = ['text'],
    template = generic_template
)
```

```python
final_prompt_template = '''Give a suitable title to the speech , and give summary in {language} in bullet points for following speech\n\n {text}'''
final_prompt = PromptTemplate(template=final_prompt_template,input_variables = ['text','language'])
```

```python
chain = load_summarize_chain(
    llm=llm,
    chain_type='map_reduce',
    map_prompt=prompt,
    combine_prompt=final_prompt,
    verbose=True
)
```

```python
chain.run({'language' : 'french','input_documents' : docs})
```
